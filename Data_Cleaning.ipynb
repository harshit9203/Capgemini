## Phone Number Column

```python
df['Phone_Number'].value_counts().head(20)
```

```python
df['Cleaned_Phone'] = df['Phone_Number'].astype(str).str.replace(r'\D', '', regex=True)
```

```python
# Extract country code (first 1-3 digits if number > 10 digits)
def get_country_code(number):
    if pd.isnull(number):
        return np.nan
    if len(number) <= 10:
        return np.nan
    return number[:-10]  # everything before last 10 digits

df['Country_Code'] = df['Cleaned_Phone'].apply(get_country_code)
```

```python
df['Country_Code'].isnull().sum()
```

```python
def get_local_number(number):
    if pd.isnull(number) or len(number) < 10:
        return np.nan
    return number[-10:]

df['Local_Number'] = df['Cleaned_Phone'].apply(get_local_number)
```

```python
df['Is_Valid_Number'] = df['Local_Number'].notnull()
```

```python
import numpy as np

def salary_format_type(s):
    s = str(s).strip()
    if s == 'nan':
        return 'Missing'
    elif any(c.isalpha() for c in s):
        return 'Contains letters (k, M, etc.)'
    elif any(c in 'â‚¹$,' for c in s):
        return 'Currency / comma'
    elif s.replace('.', '', 1).isdigit():
        return 'Numeric'
    else:
        return 'Other / invalid'

Salary_Format = salary_series.apply(salary_format_type)

# Check counts per format type
Salary_Format.value_counts()
```
