```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("bank-marketing.csv")

le = LabelEncoder()
df["marital"] = le.fit_transform(df["marital"])

df["marital"]
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

df = pd.read_csv("bank-marketing.csv")

ohe = OneHotEncoder(sparse_output=False)
encoded = ohe.fit_transform(df[["job"]])

cols = ohe.get_feature_names_out(["job"])
encoded_df = pd.DataFrame(encoded, columns=cols)

encoded
```

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder

df = pd.read_csv("bank-marketing.csv")

encoder = OrdinalEncoder(categories=[['primary', 'secondary', 'tertiary', 'unknown']])

df= encoder.fit_transform(df[['education']])

df
```
