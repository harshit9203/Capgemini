```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder

data = pd.read_csv("https://github.com/Jeevita-R/ML_datasets/raw/refs/heads/main/House_price_prediction.csv")

data
```

```python
df = pd.DataFrame(data)

df.info()
```

```python
df['Type'].unique()
```

```python
le = LabelEncoder()
df["Type_label"] = le.fit_transform(df["Type"])

df
```

```python
df_encoded = pd.get_dummies(df, drop_first=True)

df_encoded.head()
```

```python
ohe = OneHotEncoder(sparse_output=False)
encoded = ohe.fit_transform(df[["Type"]])

cols = ohe.get_feature_names_out(["Type"])
encoded_df = pd.DataFrame(encoded, columns=cols)

encoded
```

```python
encoder = OrdinalEncoder(categories=[['h', 't', 'u']])

df= encoder.fit_transform(df[['Type']])

df
```
