```python
import pandas as pd
from sklearn.preprocessing import Normalizer

data = {"Height":[162, 171, 133, 165], "Weight":[63, 60, 50, 62], "Speed":[5, 3, 7, 2]}

dF = pd.DataFrame(data, index=["S1", "S2", "S3", "S4"])
dF

normalizer = Normalizer(norm='l2')
normalized_data = normalizer.fit_transform(dF)

dF_normalized = pd.DataFrame(normalized_data, columns=dF.columns, index=dF.index)

print(dF_normalized)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

data = {"Height":[162, 171, 133, 165], "Weight": [63, 60, 50, 62], "Speed":[5, 3, 7, 2]}

dF = pd.DataFrame(data, index=["S1", "S2", "S3", "S4"])
dF

scaler = StandardScaler()
standardized_data = scaler.fit_transform(dF)

dF_standardized = pd.DataFrame(standardized_data, columns=dF.columns, index=dF.index)

print(dF_standardized)
```
