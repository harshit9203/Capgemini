```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wine = pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/winequalityN.csv')
```

```python
wine
```

```python
pop = wine['alcohol']
sample_size = 500
sample = wine['alcohol'].sample(sample_size)
sample_mean = sample.mean()
pop_mean = pop.mean()
error = pop_mean - sample_mean
```

```python
plt.scatter(range(1,sample_size+1),sample)
plt.axhline(pop_mean,color='red',label=pop_mean)
plt.axhline(sample_mean,color='blue',label=sample_mean)
plt.legend()
plt.show()
```

```python
white = wine.groupby('type').get_group('white')
red = wine.groupby('type').get_group('red')

sample_size = 50

white_sample = white['alcohol'].sample(sample_size)

white_error = pop_mean - white_sample

white_perc_error = white_error/pop_mean * 100

print(white_perc_error)



red_sample = red['alcohol'].sample(sample_size)

red_error = pop_mean - red_sample

red_perc_error = red_error/pop_mean * 100

print(red_perc_error)
```
